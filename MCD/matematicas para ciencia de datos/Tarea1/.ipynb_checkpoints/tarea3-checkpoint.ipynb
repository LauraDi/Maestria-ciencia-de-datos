# Problema 4


- Sea 1 el vector de solo unos en Rn. Para cada x ∈ Rn sea xcentered = x − µ1 donde µ es
el promedio de los elementos de x (sample mean). Responde las siguientes preguntas:

    1. Sea A centered la matriz que resulta de centrar cada columna de A. ¿Que efecto
geometrico tiene sobre A este procedimiento?

```python
df_centered = df.apply(lambda x: x-x.mean())

#view updated Array
#print(data_centered.round)
data_centered = pd.DataFrame(data=data_centered)
data_centered.round()
```

- Responde si x y y son dos columnas de A linealmente independientes, ¿tambien lo son xcentered y ycentered?

```python
# Todas las columnas de A de son independientes
print(sympy.Matrix(df.iloc[:, 0:-1].round()).nullspace())
# Todas las columnas de Acentered tambien son linealmente independientes
print(sympy.Matrix(data_centered.iloc[:, 0:-1].round()).nullspace())
```

- Responde si x y y son dos columnas de A ortogonales, ¿tambien lo son xcentered y ycentered?

- ¿Cual es la media de cualquier vector xcentered ?

```python
# la respuesta es 0 y para cada vector de A
for x in range(0, df.shape[1]):
    print(np.mean(data_centered[x]).round())
```

- Para cada x, encuentra el vector unitario con direccion xcentered

```python
# la respuesta es 0 y para cada vector de A
for x in range(0, df.shape[1]):
    print(np.mean(data_centered[x] / np.linalg.norm(data_centered[x])))
```

# Problema 5

- Encuentra, si existe, la/una combinacion lineal de las primeras 2 columnas mas cercana
a la columna numero 6. Haz lo mismo pero con las primeras 3, 4 y 5 columnas. En cada
caso decide si la respuesta existe y es unica.

```python
uno = df[0].to_numpy()
dos = df[1].to_numpy()
seis = df[5].to_numpy()
unodos = np.column_stack((uno, dos))

# Ax=B
# x = A^-1*b
pseudoInversa = np.linalg.pinv(unodos)
x = np.dot(pseudoInversa,seis)
x
```


- La respuesta existe

- Despejamos de la funcion Ax=b a x y por definicion sacamos nuestros puntos con --> x = A^-1*b y como la A es linealmente independiente tenemos una solucion unica, igual abajo.

```python
tres = df[2].to_numpy()
cuatro = df[3].to_numpy()
cinco = df[4].to_numpy()
trescuatro = np.column_stack((tres, cuatro))

# Ax=B
pseudoInversa = np.linalg.pinv(trescuatro)
x = np.dot(pseudoInversa,cinco)
x

# la respuesta existe
```
