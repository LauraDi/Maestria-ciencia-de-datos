- Encuentra, si existe, la/una combinacion lineal de las primeras 2 columnas mas cercana
a la columna numero 6. Haz lo mismo pero con las primeras 3, 4 y 5 columnas. En cada
caso decide si la respuesta existe y es unica.

```python
uno = df[0].to_numpy()
dos = df[1].to_numpy()
seis = df[5].to_numpy()
unodos = np.column_stack((uno, dos))

# Ax=B
pseudoInversa = np.linalg.pinv(unodos)
x = np.dot(pseudoInversa,seis)
x

# la respuesta existe
# Despejamos de la funcion Ax=b a x y nos queda 
```

```python
tres = df[2].to_numpy()
cuatro = df[3].to_numpy()
cinco = df[4].to_numpy()
trescuatro = np.column_stack((tres, cuatro))

# Ax=B
pseudoInversa = np.linalg.pinv(trescuatro)
x = np.dot(pseudoInversa,cinco)
x

# la respuesta existe
```
